```python
from deepqmc import Molecule
mol = Molecule(#H2O
coords=[[0.0, 0.0, 0.0], [1.4304283, 1.1071527, 0.0], [-1.4304283, 1.1071527,0.0]],
charges=[8, 1, 1],
charge=0,
spin=0,
)
print(mol)
```

Definimos a PauliNet como la red que vamos a usar y la empezamos a entrenar

```python
from deepqmc.wf.paulinet import paulinet

from deepqmc import train
from deepqmc.wf import PauliNet

net = PauliNet.from_hf(mol, cas=(5, 2)).cuda()

train(net)
```

El error que aparece es porque se detuvo luego 220 iteraciones, por ultimo evaluamos la red para ver el valor de la energía que obtenemos 

```python
from deepqmc import  evaluate
evaluate(net)
```
